```python
import os
import pdfplumber
import docx
import pytesseract
from PIL import Image, ImageEnhance, ImageFilter
from langdetect import detect
from google.colab import files

# 📌 **STEP 1: Delete Old Resumes**
def delete_old_resumes():
    for file in os.listdir("/content/"):
        if file.endswith((".pdf", ".docx", ".jpg", ".jpeg", ".png")):
            os.remove(os.path.join("/content/", file))
            print(f"🗑️ Deleted old file: {file}")

# 📌 **STEP 2: Upload Resume Files**
def upload_new_resumes():
    print("\n📂 Upload your resumes (Multiple Allowed)...")
    uploaded = files.upload()
    return list(uploaded.keys())  # Return all uploaded file names

# 📌 **STEP 3: OCR Preprocessing for Image Files**
def preprocess_image(image_path):
    image = Image.open(image_path).convert("L")  # Convert to grayscale
    image = image.filter(ImageFilter.SHARPEN)  # Sharpen the image
    return image

# 📌 **STEP 4: Extract Text from Resume**
def extract_text_from_resume(file_path):
    print(f"\n📄 Processing file: {file_path}")

    if file_path.endswith(".pdf"):
        with pdfplumber.open(file_path) as pdf:
            text = "\n".join([page.extract_text().strip() for page in pdf.pages if page.extract_text()])

    elif file_path.endswith(".docx"):
        doc = docx.Document(file_path)
        text = "\n".join([para.text.strip() for para in doc.paragraphs if para.text.strip()])

    elif file_path.endswith((".jpg", ".jpeg", ".png")):
        image = preprocess_image(file_path)  # Apply OCR Preprocessing
        text = pytesseract.image_to_string(image)

    else:
        raise ValueError(f"❌ Unsupported file format: {file_path}. Upload PDF, DOCX, JPG, or PNG.")

    # Remove extra blank lines
    clean_text = "\n".join([line for line in text.split("\n") if line.strip()])

    return clean_text

# 📌 **STEP 5: Detect Language of Extracted Text**
def detect_resume_language(text):
    try:
        return detect(text)  # Detect language
    except:
        return "Unknown"

# 📌 **RUN EVERYTHING: Delete Old, Upload New, Process, Detect Language**
delete_old_resumes()
resume_files = upload_new_resumes()

for resume_file in resume_files:
    extracted_text = extract_text_from_resume(resume_file)
    language = detect_resume_language(extracted_text)  # Detect Language

    # 📌 **Print Extracted Text and Language**
    print("\n📝 Extracted Resume Text:\n", extracted_text)
    print(f"\n🌍 Detected Language: {language}")

    # 📌 **Save Extracted Text to a File**
    output_filename = f"extracted_{resume_file}.txt"
    with open(output_filename, "w", encoding="utf-8") as f:
        f.write(extracted_text)
    print(f"✅ Extracted text saved to '{output_filename}'")
```

```python
from textblob import TextBlob#Analyze Language Proficiency (Word Count, Sentence Length, Spelling)

def analyze_language_proficiency(text):
    blob = TextBlob(text)
    word_count = len(blob.words)
    sentences = len(blob.sentences)
    avg_sentence_length = word_count / sentences if sentences else 0
    spelling_errors = sum(1 for word in blob.words if not word.lower() in TextBlob("").correct())

    proficiency_score = 100 - (spelling_errors / word_count * 100) if word_count else 0

    return {
        "word_count": word_count,
        "avg_sentence_length": avg_sentence_length,
        "spelling_errors": spelling_errors,
        "proficiency_score": round(proficiency_score, 2)
    }
```

```python
import os
import re
import PyPDF2
import docx
from collections import Counter

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as pdf_file:
        reader = PyPDF2.PdfReader(pdf_file)
        for page in reader.pages:
            text += page.extract_text() + " "
    return text.strip()

# Function to extract text from DOCX
def extract_text_from_docx(docx_path):
    doc = docx.Document(docx_path)
    return " ".join([para.text for para in doc.paragraphs])

# Function to parse resumes
def parse_resume(file_path):
    if file_path.endswith(".pdf"):
        return extract_text_from_pdf(file_path)
    elif file_path.endswith(".docx"):
        return extract_text_from_docx(file_path)
    return ""

# Function to rank candidates based on job description
def rank_candidates(resume_folder, job_description):
    job_keywords = set(re.findall(r'\b\w+\b', job_description.lower()))
    candidates = []

    for file in os.listdir(resume_folder):
        if file.endswith((".pdf", ".docx")):
            file_path = os.path.join(resume_folder, file)
            resume_text = parse_resume(file_path)
            resume_keywords = set(re.findall(r'\b\w+\b', resume_text.lower()))

            # Compute keyword match score
            match_score = len(job_keywords & resume_keywords)
            candidates.append((file, match_score))

    # Sort candidates by match score (highest first)
    candidates.sort(key=lambda x: x[1], reverse=True)
    return candidates[:5]  # Return top 5 candidates

# Job description (modify as needed)
job_desc = """
Looking for a Python Developer with experience in Flask, SQL, and machine learning.
Strong knowledge of APIs, database management, and cloud services is a plus.
"""

# Folder where resumes are stored
resume_directory = "resumes"

# Get top 5 matched candidates
top_candidates = rank_candidates(resume_directory, job_desc)

# Print results
print("\n🔍 **Top Matched Candidates:**")
for i, (file, score) in enumerate(top_candidates, 1):
    print(f"{i}. {file} - Match Score: {score}")
```

```python
import os
import re
import PyPDF2
import docx
from collections import Counter
from fuzzywuzzy import fuzz

def extract_text_from_pdf(pdf_path):
    """Extract text from a PDF file."""
    text = ""
    try:
        with open(pdf_path, "rb") as file:
            reader = PyPDF2.PdfReader(file)
            for page in reader.pages:
                extracted_text = page.extract_text()
                if extracted_text:
                    text += extracted_text + " "
    except Exception as e:
        print(f"⚠️ Error reading {pdf_path}: {e}")
    return text.lower()

def extract_text_from_docx(docx_path):
    """Extract text from a DOCX file."""
    text = ""
    try:
        doc = docx.Document(docx_path)
        for para in doc.paragraphs:
            text += para.text + " "
    except Exception as e:
        print(f"⚠️ Error reading {docx_path}: {e}")
    return text.lower()

def extract_skills_from_text(text):
    """Extract skills from resume text using regex."""
    skill_pattern = r"\b([a-zA-Z0-9\#\+\-\.]+(?:\s[a-zA-Z0-9\#\+\-\.]+)?)\b"
    words = re.findall(skill_pattern, text)
    return Counter(words)

def fuzzy_match(skill, skill_list, threshold=80):
    """Check if a skill closely matches any skill in the list using fuzzy matching."""
    for target_skill in skill_list:
        if fuzz.partial_ratio(skill.lower(), target_skill.lower()) >= threshold:
            return target_skill
    return None

def rank_candidates(resume_folder, job_description):
    """Rank candidates based on how well their resume matches the job description."""
    candidates = []

    for file in os.listdir(resume_folder):
        if file.endswith(".pdf") or file.endswith(".docx"):
            file_path = os.path.join(resume_folder, file)

            # Extract text from resume
            text = extract_text_from_pdf(file_path) if file.endswith(".pdf") else extract_text_from_docx(file_path)

            # Extract skills from resume
            resume_skills = extract_skills_from_text(text)

            # Convert extracted skills and job skills to lowercase sets
            resume_skills_set = {word.lower() for word in resume_skills.keys()}
            job_skills_set = {skill.lower() for skill in job_description["skills"]}

            # Find matching skills using fuzzy matching
            matched_skills = {
                fuzzy_match(skill, job_skills_set)
                for skill in resume_skills_set
                if fuzzy_match(skill, job_skills_set)
            }

            # Remove None values (if no match found)
            matched_skills.discard(None)

            # Calculate match score
            match_score = len(matched_skills)

            # Store results
            candidates.append((file, match_score, matched_skills))

            # Debugging: If score is 0, print extracted text
            if match_score == 0:
                print(f"\n⚠️ No skills matched in {file}. Extracted text:\n{text[:500]}...\n")

    # Sort candidates by match score
    ranked_candidates = sorted(candidates, key=lambda x: x[1], reverse=True)

    return ranked_candidates

# 🏢 **Updated Job Description**
job_desc = {
    "skills": [
        "Software Engineer", "C++", "Proofreading", "Active Listening",
        "Communication", "Critical Thinking"
    ],
    "min_experience": 0
}

# 📂 **Folder where resumes are stored**
resume_directory = "resumes"

# 🔍 **Run Candidate Ranking**
top_candidates = rank_candidates(resume_directory, job_desc)

# 📢 **Print Results**
print("\n🔍 **Top Matched Candidates:**")
for i, (name, score, skills) in enumerate(top_candidates, 1):
    print(f"{i}. {name} - Match Score: {score} - Matched Skills: {skills}")
```

```python
import openai
from googletrans import Translator
import asyncio

# Set your OpenAI API Key
openai.api_key = 'YOUR-OPENAI-API-KEY'

# Sample job postings
job_postings = {
    "Software Engineer": {
        "description": "A Software Engineer is responsible for designing, developing, and maintaining software applications.",
        "skills": ["Python", "Java", "C++", "SQL", "Problem-solving", "Teamwork"],
        "location": "Remote"
    },
    "Data Scientist": {
        "description": "A Data Scientist uses data to build models and algorithms to make data-driven decisions.",
        "skills": ["Python", "R", "SQL", "Machine Learning", "Deep Learning", "Data Visualization"],
        "location": "New York"
    }
}

# Initialize the translator
translator = Translator()

async def get_language(text):
    """Detect the language of the input text asynchronously."""
    detected_lang = await translator.detect(text)
    return detected_lang.lang

async def generate_response(prompt, language='en'):
    """Generate a response using OpenAI's API, with language handling."""

    # If the language is not English, translate the query to English
    if language != 'en':
        prompt = translator.translate(prompt, src=language, dest='en').text

    # Query OpenAI's GPT model
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=150,
        n=1,
        stop=None,
        temperature=0.7,
    )

    answer = response.choices[0].text.strip()

    # Translate the response back to the original language if needed
    if language != 'en':
        answer = translator.translate(answer, src='en', dest=language).text

    return answer

async def chatbot_response(query):
    """Handle the user query and generate an appropriate response asynchronously."""
    # Detect language asynchronously
    detected_language = await get_language(query)

    # Specific job-related queries
    if "job" in query.lower() and "software engineer" in query.lower():
        return f"Job: Software Engineer - {job_postings['Software Engineer']['description']} Required Skills: {', '.join(job_postings['Software Engineer']['skills'])} Location: {job_postings['Software Engineer']['location']}"

    elif "status" in query.lower():
        return "Your application is under review. You will be contacted for further updates."

    elif "interview" in query.lower():
        return "Your interview is scheduled for next Wednesday at 10 AM. Please be prepared."

    # If the query doesn't match specific conditions, generate an AI response
    else:
        # Generate a response in the detected language
        response = await generate_response(query, language=detected_language)
        return response

# Example Queries
async def run_queries():
    query_1 = "Can you tell me about the job of a Software Engineer?"
    query_2 = "What is the status of my application?"
    query_3 = "When is my interview scheduled?"

    print(f"Response to Query 1: {await chatbot_response(query_1)}")
    print(f"Response to Query 2: {await chatbot_response(query_2)}")
    print(f"Response to Query 3: {await chatbot_response(query_3)}")

# Run the queries asynchronously directly, without using asyncio.run()
await run_queries()
```

```python
import asyncio
from transformers import pipeline, AutoModelForQuestionAnswering, AutoTokenizer
from googletrans import Translator  # Use the normal Translator

# Initialize the Translator and Question Answering pipeline
translator = Translator()

# Use a pre-trained QA model from Hugging Face (DistilBERT for example)
model_name = "distilbert-base-uncased-distilled-squad"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
qa_pipeline = pipeline("question-answering", model=model, tokenizer=tokenizer)

async def answer_query(query: str):
    # Step 1: Detect the language of the query
    detected_language = await translator.detect(query)  # Await directly for detection
    print(f"Detected Language: {detected_language.lang}")

    # Step 2: If the query is not in English, translate it to English
    if detected_language.lang != 'en':
        translated_query = await translator.translate(query, src=detected_language.lang, dest='en')  # Translate asynchronously
        query = translated_query.text
        print(f"Translated to English: {query}")

    # Step 3: Define a context for answering the question
    context = """
    A software engineer is a professional who applies engineering principles to the design, development, maintenance, testing, and evaluation of software and systems.
    """

    # Step 4: Use the model to get an answer from the context
    result = qa_pipeline(question=query, context=context)
    answer = result['answer']
    print(f"Answer in English: {answer}")

    # Step 5: If the query was in another language, translate the answer back
    if detected_language.lang != 'en':
        translated_answer = await translator.translate(answer, src='en', dest=detected_language.lang)  # Translate asynchronously
        answer = translated_answer.text
        print(f"Answer Translated Back to {detected_language.lang}: {answer}")

    return answer

# Example Query (in any language)
query = "¿Qué hace un ingeniero de software?"  # Example in Spanish

# Run the async function inside an event loop (In Jupyter, we simply use await directly)
async def run():
    answer = await answer_query(query)
    print(f"Final Answer: {answer}")

# Use await instead of asyncio.run() in interactive environments (e.g., Jupyter)
await run()
```

```python
import re

# Example of a small job database (in real-world, this would come from a database)
job_database = [
    {
        "job_title": "Software Engineer",
        "skills_required": ["python", "java", "c++", "machine learning"]
    },
    {
        "job_title": "Data Scientist",
        "skills_required": ["python", "data analysis", "machine learning", "sql"]
    },
    {
        "job_title": "Web Developer",
        "skills_required": ["html", "css", "javascript", "react"]
    }
]

# Function to extract skills from the resume (you can enhance this with NLP techniques)
def extract_skills_from_resume(resume_text):
    # For simplicity, let's assume the resume text contains skills as comma-separated words
    # You can enhance this by using NLP to extract skills from resume sections
    skills_list = ["python", "java", "c++", "machine learning", "data analysis", "sql", "html", "css", "javascript", "react"]
    extracted_skills = []

    # Check if each skill from the list is in the resume text
    for skill in skills_list:
        if re.search(r"\b" + re.escape(skill) + r"\b", resume_text, re.IGNORECASE):
            extracted_skills.append(skill)

    return extracted_skills

# Function to recommend jobs based on the extracted skills
def recommend_jobs(resume_text):
    candidate_skills = extract_skills_from_resume(resume_text)
    print(f"Candidate Skills: {', '.join(candidate_skills)}")

    recommended_jobs = []

    for job in job_database:
        job_title = job["job_title"]
        skills_required = job["skills_required"]

        # Calculate the intersection of candidate skills and job required skills
        matching_skills = set(candidate_skills) & set(skills_required)

        if matching_skills:
            recommended_jobs.append({
                "job_title": job_title,
                "matching_skills": matching_skills
            })

    return recommended_jobs

# Main code to use the extracted resume text and recommend jobs
# Assuming extracted_text contains the extracted text from the resume
extracted_text = "John has experience in Python, Machine Learning, Data Analysis, and SQL."
recommended_jobs = recommend_jobs(extracted_text)

if recommended_jobs:
    print("\nRecommended Jobs:")
    for job in recommended_jobs:
        print(f"Job Title: {job['job_title']}, Matching Skills: {', '.join(job['matching_skills'])}")
else:
    print("\nNo matching jobs found.")
```

```python
import pytesseract
from PIL import Image
import pdfplumber

# Function to extract text from a PDF (with OCR support if needed)
def extract_text_from_pdf_with_ocr(file_path):
    text = ""
    try:
        # Try to extract text normally using pdfplumber
        with pdfplumber.open(file_path) as pdf:
            text = "\n".join([page.extract_text().strip() for page in pdf.pages if page.extract_text()])
    except Exception as e:
        print(f"Error during PDF text extraction: {e}")

    # If no text was found, apply OCR
    if not text:
        print("No text found, applying OCR...")
        with pdfplumber.open(file_path) as pdf:
            for page in pdf.pages:
                img = page.to_image()
                text += pytesseract.image_to_string(img.original)  # Apply OCR to image
    return text
```

```python
import pdfplumber
import os

def extract_text_from_pdf(pdf_path):
    """Extracts text from a given PDF resume."""
    if not os.path.exists(pdf_path):
        print(f"⚠️ Error: File '{pdf_path}' not found! Skipping...")
        return ""

    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
    return text.strip()

def enhance_resume(resume_text):
    """Enhances the extracted resume by improving structure, formatting, and highlighting skills."""
    if not resume_text:
        return "⚠️ No content extracted. Check if the PDF is scanned or empty."

    enhanced_resume = f"""
    📄 **Enhanced Resume**
    ===========================
    **👤 Profile Summary**
    {resume_text[:300]}... (Refined Summary)

    **🛠 Key Skills:**
    {'Python, Django, AI, Machine Learning' if 'Python' in resume_text else 'General IT Skills'}

    **💼 Work Experience:**
    🔹 {resume_text.split('.')[1] if len(resume_text.split('.')) > 1 else 'N/A'}
    🔹 {resume_text.split('.')[2] if len(resume_text.split('.')) > 2 else 'N/A'}

    **🏆 Achievements:**
    ✔️ {resume_text.split('.')[3] if len(resume_text.split('.')) > 3 else 'N/A'}

    **📍 Education:**
    {'Bachelor’s in Computer Science' if 'Bachelor' in resume_text else 'Education not found'}

    ===========================
    ✅ Optimized for ATS & Readability ✅
    """

    return enhanced_resume

# List of uploaded resume files
uploaded_resumes = [ "ayesha cv.pdf"]  # Ensure correct file names

enhanced_resumes = {}
for file in uploaded_resumes:
    extracted_text = extract_text_from_pdf(file)
    enhanced_resumes[file] = enhance_resume(extracted_text)
    print(f"\n📄 **Enhanced Resume for {file}**:\n", enhanced_resumes[file])
```

```python
import pdfplumber
import os
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

def extract_text_from_pdf(pdf_path):
    """Extracts text from a given PDF resume."""
    if not os.path.exists(pdf_path):
        print(f"⚠️ Error: File '{pdf_path}' not found! Skipping...")
        return ""

    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
    return text.strip()

def enhance_resume(resume_text):
    """Enhances the extracted resume by improving structure, formatting, and highlighting skills."""
    if not resume_text:
        return "⚠️ No content extracted. Check if the PDF is scanned or empty."

    enhanced_resume = f"""
    📄 **Enhanced Resume**
    ===========================
    **👤 Profile Summary**
    {resume_text[:300]}... (Refined Summary)

    **🛠 Key Skills:**
    {'Python, Django, AI, Machine Learning' if 'Python' in resume_text else 'General IT Skills'}

    **💼 Work Experience:**
    🔹 {resume_text.split('.')[1] if len(resume_text.split('.')) > 1 else 'N/A'}
    🔹 {resume_text.split('.')[2] if len(resume_text.split('.')) > 2 else 'N/A'}

    **🏆 Achievements:**
    ✔️ {resume_text.split('.')[3] if len(resume_text.split('.')) > 3 else 'N/A'}

    **📍 Education:**
    {'Bachelor’s in Computer Science' if 'Bachelor' in resume_text else 'Education not found'}

    ===========================
    ✅ Optimized for ATS & Readability ✅
    """

    return enhanced_resume

def save_resume_as_pdf(file_name, resume_text):
    """Generates a PDF from the enhanced resume text."""
    pdf_path = f"enhanced_{file_name}.pdf"

    c = canvas.Canvas(pdf_path, pagesize=letter)
    c.setFont("Helvetica", 12)

    # Write text line by line
    y_position = 750  # Start position for text
    for line in resume_text.split("\n"):
        c.drawString(50, y_position, line.strip())
        y_position -= 20  # Move down for the next line

        if y_position < 50:  # Check if space runs out on the page
            c.showPage()
            c.setFont("Helvetica", 12)
            y_position = 750  # Reset position for the new page

    c.save()
    print(f"✅ Enhanced resume saved as: {pdf_path}")

# List of uploaded resume files
uploaded_resumes = [ "ayesha cv.pdf"]  # Ensure correct file names

for file in uploaded_resumes:
    extracted_text = extract_text_from_pdf(file)
    enhanced_text = enhance_resume(extracted_text)
    save_resume_as_pdf(file.replace(".pdf", ""), enhanced_text)  # Save as PDF
```
